```python
##decoder dictionary
dec = {}  

for t in codons.split('\n'):
    k, v = t.split('\t')
    if '/' in k:
        k = k.split('/')[-1].strip()
    k = k.replace("STOP", "*")
    v = v.replace(",", "").replace(";", "").lower().replace("u", "t").split(" ")
    for vv in v:
        if vv in dec:
            print("duplicate", vv)
        dec[vv] = k
dec
```
